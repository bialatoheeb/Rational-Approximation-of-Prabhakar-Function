```python
from importnb import Notebook
import numpy as np
import matplotlib.pyplot as plt
from mittag_leffler import ml
from scipy.special import erfc
```

```python
RA = Notebook.load_file('rational_approximation_2_prabhakar_function.ipynb')
```

###### Test 1 (see [1])
- $E_{\frac{1}{2}, 1}^1(-z) = \exp(z^2)\,\text{erfc}(z) $

```python
z = np.linspace(0.1, 2., 50)
beta, gamma = 1, 1
m, n, alpha =  7, 2, 0.5
_, _, res = RA.RA(z, m, n, alpha)
sol = np.exp(z**2)*erfc(z)
assert np.allclose(res, sol, rtol=1e-03)
plt.plot(z, res, 'r*')
plt.plot(z, sol)
plt.xlabel('t')
plt.title(r'$\alpha$={}, $\beta$ = {}, $\gamma$={}'.format(alpha, beta, gamma))
plt.legend(['Rational Approx.','OPC algo.' ])
```

###### Test 2 (see [1])
- $E_{1, r}^1(-z) = \dfrac{(-1)^{r-1}}{z^{r-1}}\left\{\exp(-z) - \sum\limits_{k=0}^{r-2}\dfrac{(-z)^k}{k!}\right\} $

```python
z = np.linspace(0.1, 2., 50)
m, n, alpha, beta =  7, 2, 1, 2
_, _, res = RA.RA(z, m, n, alpha, beta)
r = beta
sol = (-1)**(r-1)/z**(r-1)*(np.exp(-z) - sum([(-z)**k/np.math.factorial(k) for k in range(r-1)]))
assert np.allclose(res, sol, rtol=1e-03)
plt.plot(z, res, 'r*')
plt.plot(z, sol)
plt.xlabel('t')
plt.title(r'$\alpha$={}, $\beta$ = {}, $\gamma$={}'.format(alpha, beta, gamma))
plt.legend(['Rational Approx.','OPC algo.'])
```

###### Test 3 (see [2])
- $E_{\alpha, \beta}^2(-z) = \dfrac{1}{\alpha}\left[E_{\alpha, \beta-1}(-z) + (1-\beta+\alpha)E_{\alpha, \beta}(-z)\right] $

```python
z = np.linspace(0.1, 2., 50) 
m, n, alpha, beta, gamma =  7, 3, 0.4, 1.7, 2
_, _, res = RA.RA(z, m, n, alpha, beta, gamma)

m, n =  7, 4   # change m, n to conform with m+n+r is even, gamma=r=1 in this case
_, _, res1 = RA.RA(z, m, n, alpha, beta-1)
_, _, res2 = RA.RA(z, m, n, alpha, beta)

sol = 1/alpha*(res1 + (1-beta+alpha)*res2)
assert np.allclose(res, sol, rtol=1e-03)
plt.plot(z, res, 'r*')
plt.plot(z, sol)
plt.xlabel('t')
plt.title(r'$\alpha$={}, $\beta$ = {} (Rational Approximations)'.format(alpha, beta))
plt.legend(['LHS','RHS' ])
```

###### Test 4 (see [2])
- $E_{\alpha, \beta}^{\gamma+1}(-z) = \dfrac{E_{\alpha, \beta-\alpha-1}^\gamma(-z) + (1-\beta +\alpha)E_{\alpha, \beta-\alpha}^\gamma(-z)}{-\alpha\gamma z}$

```python
z = np.linspace(0.1, 2., 50) 
m, n, alpha, beta, gamma =  10, 6, 0.4, 2.7, 3
_, _, res = RA.RA(z, m, n, alpha, beta, gamma+1)

n=5
_, _, res1 = RA.RA(z, m, n, alpha, beta-alpha-1, gamma)
_, _, res2 = RA.RA(z, m, n, alpha, beta-alpha, gamma)

sol = -(res1 + (1-beta+alpha)*res2)/(alpha*gamma*z)
assert np.allclose(res, sol, rtol=1e-03)
plt.plot(z, res, 'r*')
plt.plot(z, sol)
plt.xlabel('t')
plt.title(r'$\alpha$={}, $\beta$ = {} (Rational Approximations)'.format(alpha, beta))
plt.legend(['LHS','RHS' ])
```
